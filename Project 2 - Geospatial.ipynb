```python
ll = '('
ff = ')'
vv = data['Location'].tolist()
kk = []
jj = []
 
# length of substring 1 is added to
# get string from next character

for i in vv:
    if ll in str(i):
        idx1 = i.index(ll)
        idx2 = i.index(ff)
        res = i[idx1 + 1: idx2]
        kk.append(res)
        mm = i[:idx1]
        jj.append(mm)
    else:
        kk.append('')
        jj.append(i)


print(kk)
print(jj)
```
